```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

#Obtendo dados do Mercado

```python
acao = wb.get_data_yahoo('MOVI3.SA', start='2000-1-1')
```

```python
acao.head()
```

```python
acao.tail()
```

```python
acao['Adj Close'].plot(figsize=(22,8), label="PETR4")
acao['Adj Close'].rolling(7).mean().plot(label="MM7")
acao['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

```python
acao_fatiada = acao[acao.index.year >= 2015]
acao_fatiada['Adj Close'].plot(figsize=(22,8), label="PETR4")
acao_fatiada['Adj Close'].rolling(7).mean().plot(label="MM7")
acao_fatiada['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

```python
acao_fatiado = acao[(acao.index.year >= 2019)]
acao_fatiado['Adj Close'].plot(figsize=(22,8), label="PETR4")
acao_fatiado['Adj Close'].rolling(7).mean().plot(label="MM7")
acao_fatiado['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```
