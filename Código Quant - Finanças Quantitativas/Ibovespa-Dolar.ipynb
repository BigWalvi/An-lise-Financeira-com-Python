```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import pandas_datareader.data as web
import seaborn as sns
```

```python
tickers = ['^BVSP', 'USDBRL=X']
carteira = web.get_data_yahoo(tickers, start='2006-1-1')['Close']
```

```python
carteira
```

```python
carteira = carteira.dropna()
carteira
```

```python
carteira.columns = ['Ibovespa', 'Dolar']
```

```python
sns.set()
carteira.plot(subplots=True, figsize=(22,8))
```

```python
sns.heatmap(carteira.corr(), annot=True)
```

```python
carteira['Dolar'].rolling(252).corr(carteira['Ibovespa']).plot(figsize=(22,8))
```

```python
carteira['Ibov/USD$'] = carteira.Ibovespa/carteira.Dolar
carteira
```

```python
carteira['Ibov/USD$'].plot(figsize=(22,8))
```
