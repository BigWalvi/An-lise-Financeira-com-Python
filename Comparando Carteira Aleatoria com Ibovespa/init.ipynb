```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import seaborn as sns
```

```python
tickers = ["ABEV3.SA", "ITSA4.SA", "WEGE3.SA", "USIM5.SA", "VALE3.SA"]
carteira = web.get_data_yahoo(tickers, start='2015')['Adj Close']
ibov = web.get_data_yahoo("^BVSP", start='2015')['Adj Close']
```

```python
carteira
```

```python
ibov
```

```python
sns.set()
carteira.plot(figsize=(22,8))
```

```python
carteira_normalizada = (carteira/carteira.iloc[0])*100
carteira_normalizada
```

```python
carteira_normalizada.plot(figsize=(22,8))
```

```python
carteira_normalizada['Saldo'] = carteira_normalizada.sum(axis=1)
carteira_normalizada
```

```python
ibov_normalizado = (ibov/ibov.iloc[0])*500
ibov_normalizado
```

```python
carteira_normalizada["Saldo"].plot(figsize=(22,8), label='Minha Carteira')
ibov_normalizado.plot(figsize=(22,8), label='Ibovespa')
plt.legend()
```
