```python
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

#Obtendo dados do Mercado

```python
petrobras = wb.get_data_yahoo('PETR4.SA', start='2000-1-1')
```

```python
petrobras.head()
```

```python
petrobras.tail()
```

```python
petrobras['Adj Close'].plot(figsize=(22,8), label="PETR4")
petrobras['Adj Close'].rolling(21).mean().plot(label="MM21")
petrobras['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

```python
petro_fatiado = petrobras[petrobras.index.year == 2017]
petro_fatiado['Adj Close'].plot(figsize=(22,8), label="PETR4")
petro_fatiado['Adj Close'].rolling(21).mean().plot(label="MM21")
petro_fatiado['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```

```python
petro_fatiado = petrobras[(petrobras.index.year >= 2004) & (petrobras.index.year <= 2010)]
petro_fatiado['Adj Close'].plot(figsize=(22,8), label="PETR4")
petro_fatiado['Adj Close'].rolling(21).mean().plot(label="MM21")
petro_fatiado['Adj Close'].rolling(200).mean().plot(label='MM200')
plt.legend()
```
