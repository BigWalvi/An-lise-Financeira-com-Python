```
import pandas as pd
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```
empresas = ("ABEV3.SA", "AZUL4.SA", "B3SA3.SA", "BBAS3.SA", "BBDC4.SA", "BBSE3.SA", "BEEF3.SA", "BPAC11.SA", "BRAP4.SA", "BRDT3.SA", "BRFS3.SA", 
            "BRKM5.SA", "BRML3.SA", "BTOW3.SA")
```

```
compras = []
vendas = []
espera = []
for nome in empresas:
  acao = wb.get_data_yahoo(nome, start='2018')
  acao['MM21'] = acao['Adj Close'].rolling(21).mean()
  acao['MM200'] = acao['Adj Close'].rolling(200).mean()
  acao = acao.dropna()

  acao['Adj Close'].plot(figsize=(22,8), label=nome)
  acao['MM21'].plot(figsize=(22,8), label='MM21')
  acao['MM200'].plot(figsize=(22,8), label='MM200')
  plt.legend()
  plt.show()

  print(acao)
  if acao['MM21'][-1] > acao['MM200'][-1] and acao['MM21'][-2] < acao['MM200'][-2]:
    print("COMPRAR!")
    compras.append(nome)
  elif acao['MM21'][-1] < acao['MM200'][-1] and acao['MM21'][-2] > acao['MM200'][-2]:
    print("VENDER!")
    vendas.append(nome)
  else:
    print("Espera...")
    espera.append(nome)
```

```
print(f'Ordens de COMPRA para essas {len(compras)} empresas: {compras}')
print(f'Ordens de VENDA para essas {len(vendas)} empresas: {vendas}')
print(f'Nenhuma ordem para as outras {len(espera)} empresas:', end='\n')
for i in espera:
  print(f'{i}', end='\n')
```
